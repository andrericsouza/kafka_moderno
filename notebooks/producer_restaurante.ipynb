```python
from confluent_kafka import Producer
from faker import Faker
import json
import random
import uuid
import time
from datetime import datetime
```

```python
fake = Faker('pt_BR')

producer = Producer({
    'bootstrap.servers': 'kafka:29092'
})

itens = [
    'Pizza',
    'Hamburguer',
    'Batata Frita',
    'Temaki',
    'Lasanha',
    'Refrigerante'
]

status_fluxo = [
    'recebido',
    'em_preparo',
    'pronto',
    'entregue'
]

garcons = [
    'Carlos',
    'Fernanda',
    'Juliana',
    'Marcos'
]

```

```python
print('Iniciando envio de pedidos...')

while True:

    origem = random.choice([
        'cliente_app',
        'garcom_tablet'
    ])

    pedido = {
        'pedido_id': str(uuid.uuid4()),
        'origem': origem,
        'cliente': fake.first_name(),
        'garcom': random.choice(garcons),
        'mesa': random.randint(1, 20),
        'item': random.choice(itens),
        'valor': round(random.uniform(20, 150), 2),
        'status': random.choice(status_fluxo),
        'data_evento': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    }

    producer.produce(
        'pedidos_restaurante',
        value=json.dumps(pedido).encode('utf-8')
    )

    producer.flush()

    print(pedido)

    time.sleep(5)
```
