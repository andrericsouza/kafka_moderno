```python
from confluent_kafka import Consumer
import json

```

```python
consumer = Consumer({
    'bootstrap.servers': 'kafka:29092',
    'group.id': 'grupo_garcom',
    'auto.offset.reset': 'earliest'
})

consumer.subscribe(['pedidos_restaurante'])

```

```python
print('Garçom acompanhando mesas...')

while True:

    msg = consumer.poll(1.0)

    if msg is None:
        continue

    if msg.error():
        continue

    pedido = json.loads(msg.value().decode('utf-8'))

    print(
        f"Mesa {pedido['mesa']} | "
        f"Garçom: {pedido['garcom']} | "
        f"Item: {pedido['item']} | "
        f"Status: {pedido['status']}"
    )
```
